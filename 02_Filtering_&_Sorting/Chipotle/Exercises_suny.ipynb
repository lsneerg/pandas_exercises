### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
```

```python
chipo.info()
```

```python
chipo.sample(5)
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price_float'] = chipo['item_price'].str.replace("$", "").astype('float64')
```

```python
chipo.sample(5)
```

```python
cond = (chipo['quantity'] == 1) & (chipo['item_price_float'] > 10)
chipo[cond]['item_name'].nunique()
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
chipo_one_product = chipo[chipo['quantity'] == 1]
chipo_one_product
```

```python
chipo_one_product['item_name_cleaned'] = chipo_one_product['item_name'].str.replace("-", " ")
chipo_one_product
```

```python
chipo_price_per_product = chipo_one_product.groupby(by='item_name_cleaned')['item_price_float'].mean()
```

```python
chipo_price_per_product
```

### Step 6. Sort by the name of the item

```python
chipo_price_per_product
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_product = chipo_price_per_product.sort_values(ascending=False).index[0]
```

```python
cond = chipo['item_name'] == most_expensive_product
```

```python
chipo[cond]['quantity'].sum()
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl'].value_counts('order_id').sum()
```

### Step 9. How many times did someone order more than one Canned Soda?
